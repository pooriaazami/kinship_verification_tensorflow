```python
def new_name(old_path):
    new_path = old_path[:-3]
    new_path += 'jpeg'
    
    return new_path

def read_df_row(row):
    anchor = tf.train.Feature(
        bytes_list=tf.train.BytesList(value=[open(new_name(row[1][1]), 'rb').read()])
    )
    
    pos = tf.train.Feature(
        bytes_list=tf.train.BytesList(value=[open(new_name(row[1][2]), 'rb').read()])
    )
    
    neg = tf.train.Feature(
        bytes_list=tf.train.BytesList(value=[open(new_name(row[1][3]), 'rb').read()])
    )

    return tf.train.Example(
        features=tf.train.Features(feature={
            'anchor': anchor,
            'pos': pos,
            'neg': neg
        }
    ))
```

```python
def build_tf_record(name):
    df = pd.read_csv(f'data\\{name}.csv')
    record_file = f'{name}.tfrecords'

    total, _ = df.shape
    s = 0
    with tf.io.TFRecordWriter(record_file) as writer:
        for i, row in enumerate(df.iterrows()):
            ex = read_df_row(row)
            writer.write(ex.SerializeToString())

            if i % 5000 == 0 and i != 0:
                s += 5000
                writer.flush()
                print(f'5000 more triplets are done ({s} out of {total})')
```

```python
for name in names:
    print(name)
    build_tf_record(name)
```
